## Метод `.pop()`

Метод `.pop()` извлекает из списка *последний* элемент.

```python
cities = ["Москва", "Санкт-Петербург", "Новосибирск", "Калининград",
          "Владивосток", "Киев", "Москва", "Минск"]
populations = [12000000, 5300000, 1600000, 480000, 
               605000, 2900000, 12000000, 2000000]

# Извлекаем из списка последний элемент-строку, приводим его к верхнему регистру
# и присваиваем переменной city.
city = cities.pop().upper()

# Выводим данные
print(city)
print(cities)
```

```python
cities = ["Москва", "Санкт-Петербург", "Новосибирск", "Калининград",
          "Владивосток", "Киев", "Москва", "Минск"]
populations = [12000000, 5300000, 1600000, 480000, 
               605000, 2900000, 12000000, 2000000]

# Извлекаем из списка последний элемент-число, делим его на 1 000 000 
# и присваиваем переменной population 
population = populations.pop() / 1e6

# Выводим данные
print(population)
print(populations)
```

Метод `.pop()` может принимать опциональный аргумент - **индекс** элемента, который нужно извлечь.

```python
cities = ["Москва", "Санкт-Петербург", "Новосибирск", "Калининград",
          "Владивосток", "Киев", "Москва", "Минск"]
populations = [12000000, 5300000, 1600000, 480000, 
               605000, 2900000, 12000000, 2000000]

# Извлекаем из списка элемент, который находится на 6 позции.
# Делим его на 1 000 000 и присваиваем переменной population 
population = populations.pop(6) / 1e6

# Выводим данные
print(population)
print(populations)
```

## Метод `.index(value)`

С помощью метода `.index(value)` можно узнать индекс элемента по его значению (value).<br>
Если элемента в списке нет, то метод возбудит исключение `ValueError`.<br>
Если в списке содержится несколько одинаковых искомых элементов, то метод вернет **индекс первого**.

```python
cities = ["Москва", "Санкт-Петербург", "Новосибирск", "Калининград",
          "Владивосток", "Киев", "Москва", "Минск"]
populations = [12000000, 5300000, 1600000, 480000, 
               605000, 2900000, 12000000, 2000000]

# Получаем индекс элемента со значением "Москва"
moscow_index = cities.index("Москва")

# Излвекаем элемент по его индексу.
city = cities.pop(moscow_index).upper()
print(city)

# Извлекаем элемент по его индексу.
population = populations.pop(moscow_index) / 1e6
print(population)

# Выводим списки
print(cities)
print(populations)
```

Метод `.index(value, start)` может принимать необязательный второй аргумент, который сообщает, с какой позиции нужно начинать поиск. По умолчанию **start** равен нулю.

```python
cities = ["Москва", "Санкт-Петербург", "Новосибирск", "Калининград",
          "Владивосток", "Киев", "Москва", "Минск"]
populations = [12000000, 5300000, 1600000, 480000, 
               605000, 2900000, 12000000, 2000000]

# Получаем индекс элемента со значением "Москва", начинаем поиск с 1 индекса
moscow_index = cities.index("Москва", 1)

# Излвекаем элемент по его индексу.
city = cities.pop(moscow_index).upper()
print(city)

# Извлекаем элемент по его индексу.
population = populations.pop(moscow_index) / 1e6
print(population)

# Выводим списки
print(cities)
print(populations)
```

 ## Метод `.count(value)`
 
 С помощью метода `.count(value)` можно узнать сколько раз в списке встречается элемент со значением value.

```python
cities = ["Москва", "Санкт-Петербург", "Новосибирск", "Калининград",
          "Владивосток", "Киев", "Москва", "Минск"]
populations = [12000000, 5300000, 1600000, 480000, 
               605000, 2900000, 12000000, 2000000]

# Смотрим сколько раз в списке встречается элемент со значением "Москва"
cnt = cities.count("Москва")
print(cnt)

# Смотрим сколько раз в списке встречается элемент со значением "москва"
cnt = cities.count("москва")
print(cnt)
```

## Оператор `in`

С помощью оператора `in` можно узнать есть ли элемент в списке или нет.<br>
Оператор `in` возвращает логическое значение **True** или **False**.

```python
cities = ["Москва", "Санкт-Петербург", "Новосибирск", "Калининград",
          "Владивосток", "Киев", "Москва", "Минск"]
populations = [12000000, 5300000, 1600000, 480000, 
               605000, 2900000, 12000000, 2000000]

# Смотрим есть ли в списке элемент со значением "Москва"
print("Москва" in cities)

# Смотрим есть ли в списке элемент со значением "москва"
print("москва" in cities)
```
