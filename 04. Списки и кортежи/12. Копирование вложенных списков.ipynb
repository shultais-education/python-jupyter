# Копирование вложенных списков

```python
cities = [
    ["Москва", 12000000], 
    ["Санкт-Петербург", 5300000], 
    ["Новосибирск", 1600000], 
    ["Калининград", 480000], 
    ["Владивосток", 605000], 
    ["Киев", 2900000], 
    ["Минск", 2000000]
]

# Создаем копию списка
cities_copy = cities.copy()

# Выводим данные
print(cities)
print(cities_copy)
```

```python
cities = [
    ["Москва", 12000000], 
    ["Санкт-Петербург", 5300000], 
    ["Новосибирск", 1600000], 
    ["Калининград", 480000], 
    ["Владивосток", 605000], 
    ["Киев", 2900000], 
    ["Минск", 2000000]
]

# Создаем копию списка
cities_copy = cities.copy()

# Изменяем элемент во вложенном списке
cities[0][0] = "Moscow"

# Меняем целый элемент списка cities
cities[1] = ["Казань", 1200000]

# Добавлям новый элемент в список cities_copy
cities_copy.append(["Красноярск", 1000000])

# Выводим данные
print(cities)
print(cities_copy)

# Вложенный список внутри cities_copy - это тот же самый список, что в cities
print(cities_copy[0] is cities[0])
```

## Глубокое копирование с помощью `deepcopy()`

```python
# Копируем функцию deepcopy из библиотеке copy
from copy import deepcopy

cities = [
    ["Москва", 12000000], 
    ["Санкт-Петербург", 5300000], 
    ["Новосибирск", 1600000], 
    ["Калининград", 480000], 
    ["Владивосток", 605000], 
    ["Киев", 2900000], 
    ["Минск", 2000000]
]

# Создаем глубокую (полную) копию списка
cities_copy = deepcopy(cities)

# Изменяем элемент во вложенном списке
cities[0][0] = "Moscow"

# Меняем целый элемент списка cities
cities[1] = ["Казань", 1200000]

# Добавлям новый элемент в список cities_copy
cities_copy.append(["Красноярск", 1000000])

# Выводим данные
print(cities)
print(cities_copy)

# Вложенный список внутри cities_copy - это тот же самый список, что в cities
print(cities_copy[0] is cities[0])
```
