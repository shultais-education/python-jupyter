# Наследование

```python
class Auto:
    def __init__(self):
        self.number = ""  # номер автомобиля
        self.power = 0    # мощность двигателя
        
    def get_tax(self):
        """
        Возвращает налог на автомобиль в зависимости от его мощности.
        """
        
        rate = 12  # налоговая ставка
        if self.power > 100:
            rate = 25

        return rate * self.power 
        
auto = Auto()

# Устанавливаем мощность автомобиля
auto.power = 100

# Получаем налог
print("Налог (100 лс):", auto.get_tax())

# Делаем тюнинг
auto.power = 101

# Получаем новй налог
print("Налог (101 лс):", auto.get_tax())
```

## Переопределение методов

Если нам нужно изменить поведение какого-либо из методов родителя, то нам необходимо создать метод с таким же названием в подклассе. Иными словами переопределить метод.

```python
class Auto:
    def __init__(self):
        self.number = ""  # номер автомобиля
        self.power = 0    # мощность двигателя
        
    def get_tax(self):
        """
        Возвращает налог на автомобиль в зависимости от его мощности.
        """
        
        rate = 12  # налоговая ставка
        if self.power > 100:
            rate = 25

        return rate * self.power 

    
class Bus(Auto):
    
    # Переопределяем метод get_tax()
    def get_tax(self):
        """
        Возвращает налог на автобус в зависимости от его мощности.
        """        
        rate = 15  # налоговая ставка

        return rate * self.power 
        
auto = Auto()

# Устанавливаем мощность автомобиля
auto.power = 100

# Получаем налог
print("Налог автомобиля (100 лс):", auto.get_tax())

# Делаем тюнинг
auto.power = 101

# Получаем новй налог
print("Налог автомобиля (101 лс):", auto.get_tax())

# Создаём автобус
bus = Bus()

# Устанавливаем мощность автобуса
bus.power = 300

# Получаем налог
# Будем вызван метод .get_tax() из класса Bus
print("Налог автобуса (300 лс):", bus.get_tax())
```
